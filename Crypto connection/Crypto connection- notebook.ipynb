### Upon analyzing, I found that bitcoin leads ethereum by 24 days.
#### Now, I need to apply trading strategy which would exploit this lead-lag relationship. To do so, the strategy I would be using is MACD.

Here, I defined a function for MACD strategy, and calculated macd line and signal line for ethereum.

```python
def get_macd(price,slow,fast,smooth):
    exp1 = price.ewm(span = fast, adjust = False).mean()
    exp2 = price.ewm(span = slow, adjust = False).mean()
    macd = pd.DataFrame(exp1-exp2).rename(columns = {'Close':'macd'})
    signal = pd.DataFrame(macd.ewm(span = smooth, adjust = False).mean()).rename(columns = {'macd':'signal'})
    frames =  [macd, signal]
    df = pd.concat(frames, join = 'inner', axis = 1)
    return df

eth_macd=get_macd(eth.Close,26,12,9)
eth_macd
```

```python
macd_buy_signals=[]
macd_sell_signals=[]
macd_signals=[]

macd_status=''
for i in range(len(eth)):
    if eth_macd['macd'][i] < eth_macd['signal'][i] and macd_status != 'SELL':
        macd_sell_signals.append(i)
        macd_signals.append(('SELL'))
        macd_status = 'SELL'
    elif eth_macd['macd'][i] > eth_macd['signal'][i] and macd_status != 'BUY':
        macd_buy_signals.append(i)
        macd_signals.append(('BUY'))
        macd_status = 'BUY'
    else:
        macd_signals.append(('NONE'))

print(macd_sell_signals)
print(macd_buy_signals)
print(macd_signals)
```

```python
plt.figure(figsize=(10, 3))
plt.plot(eth.index, eth['Close'], label='Eth Price')
eth_macd_buy=eth[eth_macd.macd>eth_macd.signal]
plt.scatter(eth_macd_buy.index, eth_macd_buy['Close'], marker='^',color='green', label='buy signals')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('macd buy signals - ETH')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 3))
plt.plot(eth.index, eth['Close'], label='Eth Price')
eth_macd_buy=eth[eth_macd.macd<eth_macd.signal]
plt.scatter(eth_macd_buy.index, eth_macd_buy['Close'], marker='v',color='red', label='sell signals')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('macd sell signals - ETH')
plt.legend()
plt.show()
```

```python
macd_investment=100
macd_number_of_stocks=macd_investment/(eth['Close'][0])
macd_in_hand = macd_investment - macd_number_of_stocks * eth['Close'][0]
macd_return=[]


for i in range(len(eth)):
    if macd_signals[i]=='BUY':
        temp = macd_in_hand / eth['Close'][i]
        macd_in_hand -= temp * eth['Close'][i]
        macd_number_of_stocks += temp
    elif macd_signals[i]=='SELL':
        macd_in_hand += macd_number_of_stocks * eth['Close'][i]
        macd_number_of_stocks = 0
    
    macd_return.append(macd_in_hand + macd_number_of_stocks * eth['Close'][i])


macd_returns=macd_in_hand + macd_number_of_stocks * eth['Close'][-1]
print("The return on investing 100000 rupees is:",macd_returns)

macd_percentage_returns=format((((macd_returns)-(macd_investment))/macd_investment)*(100),".2f")
print("The percentage return in case of macd is:",macd_percentage_returns,"%")
#returns
```

Here, I implemented MACD strategy on bitcoin, and will use these signals for lead-lag analysis.

```python
bitcoin_macd=get_macd(bitcoin.Close,26,12,9)
bitcoin_buy_signals=[]
bitcoin_sell_signals=[]
bitcoin_signals=[]

bitcoin_status=''
for i in range(len(bitcoin)):
    if bitcoin_macd['macd'][i] < bitcoin_macd['signal'][i] and bitcoin_status != 'SELL':
        bitcoin_sell_signals.append(i)
        bitcoin_signals.append(('SELL'))
        bitcoin_status = 'SELL'
    elif bitcoin_macd['macd'][i] > bitcoin_macd['signal'][i] and bitcoin_status != 'BUY':
        bitcoin_buy_signals.append(i)
        bitcoin_signals.append(('BUY'))
        bitcoin_status = 'BUY'
    else:
        bitcoin_signals.append(('NONE'))

print(bitcoin_sell_signals)
print(bitcoin_buy_signals)
print(bitcoin_signals)
```

```python
bitcoin_df=pd.DataFrame(bitcoin_signals).rename(columns = {0:'bitcoin_signals'})
eth_df=bitcoin_df.shift(24).rename(columns={'bitcoin_signals':'eth_signals'})
```

```python
lead_lag_investment=100
lead_lag_number_of_stocks=lead_lag_investment/(eth['Close'][0])
lead_lag_in_hand = lead_lag_investment - lead_lag_number_of_stocks * eth['Close'][0]
lead_lag_return=[]

for i in range(len(eth_df)):
    if eth_df['eth_signals'][i]=='BUY':
        temp = lead_lag_in_hand / eth['Close'][i]
        lead_lag_in_hand -= temp * eth['Close'][i]
        lead_lag_number_of_stocks += temp
    elif eth_df['eth_signals'][i]=='SELL':
        lead_lag_in_hand += lead_lag_number_of_stocks * eth['Close'][i]
        lead_lag_number_of_stocks = 0

    lead_lag_return.append(lead_lag_in_hand + lead_lag_number_of_stocks * eth['Close'][i])

lead_lag_returns=lead_lag_in_hand + lead_lag_number_of_stocks * eth['Close'][-1]
print("The return on investing 100000 rupees is:",lead_lag_returns)

lead_lag_percentage_returns=format((((lead_lag_returns)-(lead_lag_investment))/lead_lag_investment)*(100),".2f")
print("The percentage return in case of lead-lag analysis is:",lead_lag_percentage_returns,"%")
```
